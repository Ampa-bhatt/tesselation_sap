In today’s highly competitive retail in-store industry, it is crucial to recognize that efficient
management of store operations is essential, to achieve and sustain customer satisfaction, as well
as cost-effectiveness. Hence one of the major factors which distinguishes a successful retailer from

# I. INTRODUCTION

Measuring operational efficiency is crucial for the success of a business. It helps decision makers to identify areas for improvement, increase productivity, improve the customer experience and gain a competitive advantage.

* Better Customer Experience: By measuring operational efficiency, retailers can assess the speed, accuracy and quality of customer service, and identify areas for improvement. This can help to enhance the customer experience, increase customer satisfaction and build customer loyalty.

* Inventory Management: Measuring operational efficiency can help retailers to better manage their inventory levels and reduce waste. This can help to reduce the cost of carrying inventory, improve the accuracy of inventory counts and reduce the risk of stock shortages.

* Supply Chain Management: Retailers can use operational efficiency metrics to monitor the performance of their supply chain partners and identify areas for improvement. This can help to improve the reliability and speed of the supply chain, reduce costs and increase profitability.


However, there are certain key challenges in measuring efficiency, including but not limited to:

* Data Availability: One of the biggest challenges in measuring operational efficiency is the availability and quality of data. Retailers may have difficulty collecting accurate and comprehensive data from different departments and systems, making it challenging to make informed decisions.

* Complex Processes: Retail operations can be complex, involving multiple departments, systems and processes. Measuring the efficiency of these processes can be challenging, as it may require detailed analysis and a thorough understanding of the processes involved.

* Subjectivity: Measuring operational efficiency can be subjective, as different individuals may have different interpretations of what constitutes efficiency. This can make it difficult to develop meaningful metrics and compare performance across different retail operations.

* Resistance to Change: Change can be challenging, and many retailers may resist implementing changes that are necessary for improving operational efficiency. This can make it difficult to implement new processes, technologies or systems that are designed to increase efficiency.


* Lack of Standardization: There is a lack of standardization in the retail sector when it comes to measuring operational efficiency. This can make it challenging to compare performance across different retailers and to make informed decisions based on performance metrics.

This paper presents a case study of a combined application of computational geometry (Voronoi Tessellation) and economic modeling (Data Envelopment Analysis) techniques to determine efficiency of retail stores based on a combination of demographic and operation's data, keeping in mind the aforementioned challenges, and therefore seeks to present decision makers with insights for determining efficient and inefficient units in their business, and therefore take remedial measures.

## i. Voronoi Tessellation

Voronoi Tessellation and tiled regions are related concepts in spatial analysis. Voronoi Tessellation refers to a way of dividing a plane into regions based on the distance between a set of seed points, while tiled regions refer to the resulting partitions created by the Voronoi Tessellation.

Voronoi Tessellation is a mathematical concept that was first introduced by the Ukrainian mathematician Georgy Voronoi in 1908. Given a set of seed points in a plane, the Voronoi Tessellation partitions the plane into polygonal cells, with each cell being associated with a unique seed point. The cell of a seed point consists of all points in the plane that are closer to that seed point than to any other seed point. In other words, the Voronoi Tessellation creates a mapping between each seed point and the region of the plane closest to it.

The resulting tiled regions of the Voronoi Tessellation are often referred to as Voronoi diagrams. These diagrams are used in a variety of fields, including computer science, geography, and engineering. For example, they can be used to model the distribution of resources, such as water or food, in a landscape, or to understand the patterns of communication between individuals in a social network.

Thus, Voronoi Tessellation and tiled regions are powerful tools for understanding the spatial relationships between points in a plane. By dividing a plane into regions based on distance, it is possible to gain insights into complex systems and make informed decisions based on the resulting information.

Mathematically, Voronoi tessellation can be defined as follows:

Let X be a set of n points in a d-dimensional space. For each point xi in X, the Voronoi cell V(xi) is defined as the set of all points in the space that are closer to xi than to any other point in X:

V(xi) = {y in d-dimensional space | dist(y, xi) <= dist(y, xj), for all j != i}

Where:

d = the dimension of the space

X = the set of n input points

xi = a point in X

xj = another point in X, where j != i

dist(y, xi) = the Euclidean distance between the points y and xi

The Voronoi tessellation can be visualized as a partition of the space into cells, with each cell representing the Voronoi cell of a single point in X.

In summary, Voronoi tessellation is a mathematical concept that partitions a space into regions based on the proximity of points in the space. The Voronoi cell of a point xi is defined as the set of all points in the space that are closer to xi than to any other point in X.

## ii. Data Envelopment Analysis

Data envelopment analysis (DEA) is a non-parametric programming method that has been widely used in the retail sector for measuring the efficiency of decision-making units (DMUs). It provides a framework for comparing the performance of multiple DMUs based on a set of inputs and outputs, and enables decision-makers to identify best practices and areas for improvement. DEA has been applied in various aspects of retail operations, including store management, supply chain management, and customer service, among others.

The output-oriented radial CCR (Charnes, Cooper, and Rhodes) model is a variant of the radial CCR model in data envelopment analysis (DEA) that is used to evaluate the relative efficiency of decision making units (DMUs) based on their outputs. This model is particularly useful when the outputs are the primary focus of the analysis, as opposed to the inputs.

The output-oriented radial CCR model is formulated as follows:

Maximize:

Z = λ

Subject to:

Σ(xi) = 1 (for all i = 1, 2, ..., n)

yi ≥ λxi (for all j = 1, 2, ..., m)

xij ≥ 0 (for all i = 1, 2, ..., n and j = 1, 2, ..., m)

Where:

Z = the objective function

λ = a scalar value that represents the efficiency of the DMU

n = the number of DMUs

m = the number of inputs or outputs

xi = the weight assigned to the jth input or output for the ith DMU

yj = the amount of the jth output produced by the ith DMU

The output-oriented radial CCR model evaluates the efficiency of each DMU by maximizing the value of λ subject to the constraints. The value of λ represents the efficiency score of each DMU and is determined by dividing the sum of the outputs by the sum of the inputs. If the value of λ is equal to 1, it means that the DMU is efficient, and if it is greater than 1, it means that the DMU is efficient with an increase in scale.

In summary, the output-oriented radial CCR model is used to evaluate the relative efficiency of a set of DMUs by considering the outputs as the primary focus and taking into account the non-linear nature of the production process. The efficiency of each DMU is evaluated by maximizing the value of λ subject to constraints.

### g. Summary table

Following is a summary table showing each catchment area along with its respective target population. The target population is obtained from the mapped zipcodes and the corresponding population from the census data along with few filtration on age group.
```python
def create_dmu_inputs(cencus_df,polygon_df):
  
  L=[]
  for i in polygon_df.iterrows():
    pip_mask = cencus_df.geometry.within(polygon_df.loc[i[0],'geometry'])
    pip_data = cencus_df.loc[pip_mask]
    total_population_per_region=pip_data['population'].sum()
    total_density=pip_data['density'].sum()
    tile_region=i[-1]
    L.append((i[0],total_population_per_region,total_density,tile_region))

  return L

data=create_dmu_inputs(us_data_points,supermarkets_voronoi)
data
data_out=pd.DataFrame(data=data,columns=["Tile_ID","Population","Density","Tile_Region"])
data_out["DMU_Name"]=[f"{row['Store_ID']}-{row['addr:city']}" for index,row in supermarkets.iterrows()]
data_out=data_out.filter(["Tile_ID","DMU_Name","Population","Density","Tile_Region"])
data_out.head()
```

### b. Scenario methodology

The analysis is broken up into three scenarios:

* Scenario 1 deals with ***Floor_Space*** as the only input. Outputs used are ***Sales*** and ***Customers***.

* Scenario 2 uses ***TAM*** (obtained from Voronoi tessellation and filtered as discussed in section IV.g.) as the only input. Outputs remain the same as in scenario 1.

### c. Scenario 1: Floor space input vs. outputs

In the first scenario, we measure the technical efficiency across DMUs for the given ***Floor_Space***. The output variables considered here are ***Sales*** per day (value) and ***Customers*** per day (unique) against which the efficiency of the DMUs can be determined. Once the efficiency values are determined via the radial CCR model, the frontier is plotted accordingly.

Now the following table is derived from the table shown in section IV (Implementation) subsection iii -> a [Data Envelopment Analysis -> DEA asset data]

Here, for the first row in the table below, the store (with index 0) had floor space of 14 units, sales of 7 units and customers of 11 units from the earlier table as referred above. So, customers per floor space was computed by dividing 11 units of cutomers by 14 units of floor space. The result is 0.785.

Similarly, sales per floor space was computed by dividing 7 units of sales by 14 units of floor space. The result is 0.5.

```python
customers_vs_sales_fs = pd.DataFrame()
customers_vs_sales_fs["Store"] = data["DMU_Name"]
customers_vs_sales_fs["Customers/FS"] = data["Customers"]/data["Floor_Space"]
customers_vs_sales_fs["Sales/FS"] = data["Sales"]/data["Floor_Space"]
customers_vs_sales_fs.head()
```

```python
# plot the line plot of efficent and a scatter plot of inefficient in the same plot

plt.figure(figsize=(10, 10))
fig, ax = plt.subplots()

customers_vs_sales_fs.plot(ax=ax, y='Sales/FS', x='Customers/FS', kind='scatter', color='red')
plt.ylabel('Sales/FS')
plt.xlabel('Customers/FS')
plt.title('Scatter plot')
plt.show()
```

In the above scatter plot, both the axes are measuring  two different kinds of outputs (Sales in Y axis and Customers in X axis), per unit of one input(Floor Space). Hence the X axis represents number of customers per unit floor space, while the Y axis represents number of sales per unit of floor space. We would always like our X and Y values to be maximum per unit of specific input measure taken into consideration for the above graph. Hence, if we connect more than two points, those having the most high values in combination of X and Y measures, it will form a curve which is concave (with respect to the origin) in nature (it is an exception if more than 2 points comes in a straight line).

Hence, for maximizing outputs versus inputs, we would always prefer the curve which represents maximum efficiency to be concave (with respect to origin) in nature. This curve is called as "frontier curve" for output-oriented analysis considering two outcomes versus one input measure. A convex curve (with respect to the origin) will not be preferred as it will have points which are less efficient than desired ideally.

```python
# convert data to array
output_array_fs = np.array(data[["Sales","Customers"]])
input_array_fs = np.array(data[["Floor_Space"]])

# define and solve the DEA radial model
model_fs = DEA.DEA(output_array_fs, input_array_fs, rts=RTS_CRS, orient=ORIENT_OO, yref=None, xref=None)
model_fs.optimize('nshntmlhtr@gmail.com')

# get the technical efficiency on a scale of 0 to 1
theta_fs=pd.DataFrame(1/model_fs.get_theta(),columns=["Technical Efficiency"])
theta_fs
```

In the above table, following the radial CCR output-oriented DEA model, efficiencies were computed for each of the 30 stores. 5 stores have highest efficiency of greater than or equal to 0.8 and others have opportunities for efficiency improvement.

```python
customers_vs_sales_fs["Efficiency"] = theta_fs
customers_vs_sales_fs.head()
```

The above table is a summary table for customers per floor space, sales per floor space and the efficiency computed previously.

```python
customers_vs_sales_efficient_fs = customers_vs_sales_fs[customers_vs_sales_fs['Efficiency'] == 1.0]
customers_vs_sales_efficient_fs = customers_vs_sales_efficient_fs.sort_values(by=["Sales/FS"],ascending=False)
customers_vs_sales_efficient_fs
```

The above table lists the efficient DMUs having efficiency equal to 1.

```python
customers_vs_sales_inefficient_fs = customers_vs_sales_fs[customers_vs_sales_fs['Efficiency'] < 1.0]
customers_vs_sales_inefficient_fs
```

The above table lists the inefficient DMUs i.e. DMUs having efficiency less than 1.

```python
plt.figure(figsize=(10, 10))
fig, ax = plt.subplots()
customers_vs_sales_efficient_fs.plot(ax=ax, y='Sales/FS', x='Customers/FS', kind='line', color='gray')
customers_vs_sales_efficient_fs.plot(ax=ax, y='Sales/FS', x='Customers/FS', kind='scatter', color='green', marker = 'x')
customers_vs_sales_inefficient_fs.plot(ax=ax, y='Sales/FS', x='Customers/FS', kind='scatter', color='red')

plt.ylabel('Sales/FS')
plt.xlabel('Customers/FS')
plt.title('DMU Comparison')
plt.show()
```

The above graph shows the frontier curve plotted by joining the data points representing the efficient DMUs from the above tables  in Scenario 1. Considering, that efficiency in this case relates to output maximization, the frontier curve is concave (with respect to origin) in nature as discussed peviously.

Hence, as mentioned before, if we connect more than two points with the higher combination of values for X and Y axis, we would probably get a curve that is again concave (with respect to orginin) in nature. This would be the "frontier efficiency" curve.

It is to be noted that since we are maximizing outputs, we would expect the frontier curve to be concave (with respect to orgin) in nature. However, if we were minimizing inputs against fixed output to achieve better efficiency, the frontier curve would have preferably been convex (with respect to origin) in nature.

```python
# convert data to array
output_array_TAM = np.array(data[["Sales","Customers"]])
input_array_TAM = np.array(data[["TAM"]])

# define and solve the DEA radial model
model_TAM = DEA.DEA(output_array_TAM, input_array_TAM, rts=RTS_CRS, orient=ORIENT_OO, yref=None, xref=None)
model_TAM.optimize('nshntmlhtr@gmail.com')

# get the technical efficiency on a scale of 0 to 1
theta_TAM=pd.DataFrame(1/model_TAM.get_theta(),columns=["Technical Efficiency"])
theta_TAM
```

In the above table, following the radial CCR output-oriented DEA model, efficiencies were computed for each of the 30 stores. 3 stores have highest efficiency of 1.0 and others have opportunities for efficiency improvement.

## i. Findings: Tessellation & TAM

Total addressable market (TAM) is important for retail store performance evaluation because it helps the store understand the potential size of the market and the level of demand for their products or services. By knowing the TAM, the store can estimate their market share and identify opportunities for growth. For example, if the store is only capturing a small portion of the TAM, there may be opportunities to increase sales by expanding product offerings or targeting new customer segments.

However, estimating TAM may prove challengiing for the following reasons:

1. Difficulty defining the market: The definition of the market can be vague or subjective, making it challenging to determine the TAM. The entire population will not be the target market. For a generic multi-brand retail chain, age group consideration is definitely critical.

2. Market dynamics: The market can be constantly changing, with new competitors and shifting customer preferences, making it difficult to accurately estimate the TAM.

3. Limited resources: Estimating the TAM can require significant time and resources, which may be limited for smaller businesses.

Keeping this in mind, an attempt has been made to gauge the probable TAM from publicly available census data by using Voronoi tessellation to estimate the catchment areas of each store to provide a rough geographic estimate of TAM popualton for each store after applying age filters.

The aim was to understand store performance based on the above TAM population when combined with regular operational factors such as store floor space, in order to get a holistic picture of store efficiency and address slacks.

# VI. CONCLUSION

## a. Composite index for multiple inputs and outputs

Multiple input and output factors should be taken into consideration whenever  discussing the efficiency of business units. As demonstrated via the experiments conducted in this case, it is necessary to form a composite index of all operational performance metrics for a holistic view of unit performance. This enables in complimentary analysis of individual efficiency measures along with composite measures which leads to better management decisions.

## b. Some external factors like floor space cannot be directly manipulated

However, it is to be noted that some factors like floor area and total target addressable market (TAM) may not be as easily optimized as other factors. Therefore the opportunity lies in maximizing total output for the fixed input metrics. Hence, it is important to decide which model is more relevant - maximising outputs or minimising inputs to compare and benchmark efficiency.

## c. Use of output maximization model for fixed inputs

As demonstrated in this paper, operational efficiency based on ouput maximization via Data Envelopment Analysis (DEA) following identification of key input metrics like total addressable market obtained via tessellation using census and store data can lead to various important insights that can be used to plan a course of action to stay competitive and lean in the market. Thus, using tessellation to determine TAM helps in providing an input factor which may not be always directly available for similar business cases while deploying DEA.

## d. Visual inferences from combo charts

Visual insights obtained from plotting sales and efficiency with respect to floor space and total addressable market respectively lead to interesting data discoveries pertaining to the changes in efficiency and sales based on the input metrics. This provdes better explainability by supplementing the results from DEA with actionable items for further consideration and analysis.

## e. Further opportunties

As with any case study using a combination of algorithms and certain available dimensions, this paper does not detail a few methods and doesn't include certain other considerations that can add more value from the perspective of outcome inference. Dimension priority methods can be deployed using either regression, pricipal component analysis or analytical hierarchical process (AHP) as any business will have many inputs and outputs measures and it is important to consider vital few instead of trivial many. Buyers of any retail outlet can live in a semi-urban locality and commute to office in an urban locality and visit rural areas within a state very frequently. So, TAM determination is not easy and never precise. Fuzzy set theory can be applied to DEA models to handle imprecise inputs for considering these situations by defining tolerance levels on the objective function and constraint violations. Also, competitor information plays a vital role in decision making while comparing and benchmarking retail store efficiency. Public data (with available competitor information on stores) can be accessed more comprehensively to account for the same to further the robustness of the analysis outcomes.
